```python
%load_ext autoreload
%autoreload 2

import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

sns.set_style('whitegrid')
```

```python
n_samples = 10000
k = 5
v = 1

samples = []
tau = np.random.gamma(shape=k/2, scale=2/(k*v))
x = np.random.normal(loc=0, scale=np.sqrt(1/tau))
for epoch in range(n_samples):
    tau = np.random.gamma(shape=(k+1)/2, scale=2/(k*v+x**2))
    x = np.random.normal(loc=0, scale=np.sqrt(1/tau))
    samples.append((x, tau))
```

```python
df = pd.DataFrame(samples, columns=['x', 'tau'])
```

```python
df.x.plot(figsize=(20, 4))
```

```python
plt.scatter(df.x, df.x.shift(), s=3.5)
```

```python
df.tau.plot(figsize=(20, 4))
```
