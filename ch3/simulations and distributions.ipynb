```python
%load_ext autoreload
%autoreload 2

import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

sns.set_style('whitegrid')
```

```python
gs = np.random.gamma(shape=[2,20], scale=1, size=(100000, 2))
beta_from_gammas = gs[:,0] / gs.sum(1)
beta_direct = np.random.beta(2, 20, size=100000)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4.5))

sns.kdeplot(gs, ax=ax[0])
sns.kdeplot(beta_from_gammas, ax=ax[1], label='From Gammas')
sns.kdeplot(beta_direct, ax=ax[1], label='Beta Direct')

ax[0].legend(['Ga1', 'Ga2'])
ax[1].legend(loc='best')

fig.tight_layout()
fig.show()
```
