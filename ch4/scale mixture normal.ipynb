```python
%load_ext autoreload
%autoreload 2

import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats

sns.set_style('whitegrid')
```

```python
v = 10
s = 5

# precision samples at 1 unit precision with v degree of freedom
λs = np.random.gamma(shape=v/2, scale=2/v, size=10000)
# conditional normal -> T
xs = np.random.normal(loc=0, scale=np.sqrt(s * 1/λs))

# compare it with the actual T-distribution samples
x_range = np.linspace(np.min(xs), np.max(xs), 10000)
ts = stats.t(df=v, loc=0, scale=np.sqrt(s))
```

```python
sns.kdeplot(xs)
plt.plot(x_range, ts.pdf(x_range))
```
